# 数据监测
一般一个广告投放的整体流程是：曝光->点击->到达->站内活动及跳转->转化->产品购买，其中各个环节都会安置监测代码来收集数据。

- 根据收集到的数据，同原定的排期计划做对比，来评估不同媒介渠道的广告投放真实性及效率效果。
- 通过不同环节间的转化率来评价不同媒介渠道，分析各渠道带来的流量质量及对营销目标的贡献率。

在数据对比及确定结算依据时，广告主及广告投放甲乙双方，各自都会以自己的数据说话，这自然也就需要一个相对**第三方的监测数据分析方**，其对上述那些广告投放整体流程中的各个环节的数据进行收集，并作为双方都能认可的结算依据。

## TA%浓度KPI
品牌广告投放中的一个重要KPI考核指标就是TA浓度（TA%）。TA%的简单理解可以是一次广告投放下来，有多少次广告曝光被目标人群收看，即目标人群的广告曝光数占总曝光数的比率。

### IGRP
IGRP由于OTV经常需要配合补充电视广告在目标城市及目标人群市场的投放，所以沿用了GRP相关的概念：

```
IGRP(%) = Reach(目标人群到达率) * Average Frequency(目标人群收看广告的平均次数) * 100
```

以下用一个例子来复习下如何计算相关指标

- TA Universe(Universe): 即全部符合条件的目标人数。假如A市网民有2000万，其中18~35岁男网民（Target Audiency）人口占25%，A市TA Universe则为500万。
- TA PV(TA Page Views): 即符合投放条件的广告展示量。假设投放的广告展示量为3000万次，这3000万次中，符合"18~35""男"目标人群要求的网民广告展示量为750万，即TA PV为750万。
- TA UV(TA Unique Views): 即符合投放条件的独立广告UV量（可理解为广告触达到的目标用户数）。假设750万的TA PV中，不重复的独立UV量（TA UV）为50万。
- Average Frequency(Frequency，目标人群收看广告的平均次数，其中匹数又称频次): 即TA PV除以TA UV的值。假设TA PV为750万，TA UV为50万，则750万除以50万等于15，所以：
- Reach: 某个地区看到广告的目标受众占比该地区所有目标受众的比例。即Reach = (Ta UV)/(TA Universe) * 100% = 50万/500万 = 10%
- IGRP(%) = Reach * Average Frequency * 100 = 0.1 * 15 * 100 = 150

### TA N+ Reach
由于广告收看的频次会影响目标受众，频次太低影响力就不够，就是我们常说的"强度"不够，所以会有广告主关心TA N+ Reach。N+: 广告曝光N次及N次以上。

TA N+ Reach: 收看广告"N次及N次以上"的目标人群到达率。 := (N+ TA UV) / (TA Universe)

### TA%(TA浓度):
TA浓度一般以百分比形式出现，代表目标受众用户的曝光数占总稳定曝光数的比。:= (TA PV)/(Stable PV)

### 注意事项
一般广告投放最终的受众触达情况(iGRP, TA%)，广告主方还是以第三方监测提供的数据为主要参考依据。然后问题恰恰就出在这个地方，虽然互联网数字化高速发展，然而各方并没有形成相对一致的评估受众的机制或标准库。

第三方监测是小样本库（秒针叫Panel库）推及的方式，来推算TA覆盖率。大体的计算过程如下（以某个地域为例）：
1. 用本次广告曝光的人群ID包，同第三方监测样本库中目标TA样本人群ID包，匹配出第三方监测样本库命中的人群ID数作为分子。
2. 用本次广告曝光的人群ID包，同第三方监测样本库中所有样本人群ID包，匹配出第三方监测样本库命中的人群ID数作为分母。
3. 用分子/分母乘以该地域总曝光人群UV数，得出各地域覆盖的总TA人数。
4. 再用该总TA人数除以该地域的目标人口基数得到TA Reach数

而是广告投放过程中，广告投放执行方并不是采用同第三方一致的方式来指导投放的。正由于这样的标准不一致，海量行为分析同小样本推及，导致了视频广告投放的效果评估成为现在业内的一个重大难题，除非广告投放方向第三方监测的样本库看齐。

PS：秒针目前已经关闭了其样本库接口，也就是想要向其看齐而不可得，品牌广告主目前又大量以秒针的样本库为基准，TA% KPI目前是投放中的一大难题。

## 监测数据GAP
### 地域GAP
GAP即差距、差异，常常指广告投放方观测到的数据同第三方监测或广告主观测到数据的差异。由于这个差距会涉及到广告结算，是各方都十分敏感的问题。

广告主给媒体结算，一般都是按照第三方监测的数据结果进行的。而媒体监测到的不同地域的曝光量，同第三方监测到的曝光量存在一定的GAP，这种GAP就是俗称的地域GAP。

广告主要求的地域定向，往往就是按不同地域、不同的价格及完成量来结算的，一般不会不分地域而按总量来结算，尤其是在视频广告领域。如果某地域没有按排期投够，就需要补量。而投多了广告主也不会多出钱。所以各方都十分关注且特别紧张地域GAP。

地域GAP的成因：1. 因策略路由导致的（我们常称之为IP漂移）。2. 各方IP库不同，或者IP库同步不一致。3. 运营执行配置失误，造成地域定向出错。

#### 如何缓解地域GAP
- 广协IP库，以及各方约定同步日期，基本上解决了第二点原因
- 投放执行方通过培训，监控数据可尽量避免第三点这种低级失误
- 媒体方通第三方监测建立同源机房，适度缓解地域GAP。也就是针对不同的媒体监测代码，采用该媒体特有的域名。这样通过域名DNS指向的方式，直接将该媒体用户的监测请求指向与媒体同在一个机房的，即同源机房的第三方监测服务器。
    + 同源机房代码目前仅对视频媒体适用，对大部分的中小媒体不适用。

### 结算方式的坑
ADX卖方同买方（DSP）结算之间是按广告竞价成功金额进行结算的，而广告主同DSP方一般是以第三方监测数据为依据来结算的，所以DSP方需要承担这个GAP。一般这个GAP为10%左右算正常。往往客户有地域定向的要求，这就造成DSP需要承担更多的GAP风险。

### PDB中的损耗问题
传统采买中广告主、媒体方同第三方监测基本达成共识，若媒体放数据同第三方监测数据GAP小于10%，媒体方能接受按第三方监测来结算。但PDB系统的加入增加了中间环节，会导致这个GAP有增大的可能性。

对于这个问题一般有以下两种解决方法
- 技术对接要多做真实流量的投放测试，观测出这个GAP是否在可接受范围内。曝光GAP，点击GAP都是观测的重点。
- 媒体端增加一个打底广告，并加上打底广告监测代码，若PDB方响应超时，媒体端可以展示打底广告。

### 频次监测问题
目前真实情况是大部分媒体都是按广协统一的标准处理设备ID，不过实际投放中依旧会出现投放方与第三方监测的收集的设备不一致，其中带来的最大的问题现象就是超频。

特别是MMA的SDK监测方式中，投放方与第三方监测的设备ID收集不一致问题更为严重。

## 第三方监测
以下先介绍几家合作较多的第三方监测公司
- 秒针(MiaoZhen Systems): PC，MOB均可监测，主要监测前段数据（曝光，点击，可见性），并且给广告主出具TA，IGRP等监测报告，以及GIVT，SIVT质量评估报告。收购合并AdMaster，所以AdMaster就不再介绍了。
- Sizmek：国际品牌广告使用较多。主要特点是可见性，品牌安全等监测。
- 尼尔森：类似秒针，只是国内市场份额没有秒针大。
- 国双科技：监测数据主要涉及数字营销、新媒体、电信运营商、电子政务、司法，汽车领域等，垂直类广告主会使用它。

### MMA监测标准（以秒针为例）
类似广告对接，第三方监测也分SDK，API两种方式
- SDK：媒体方需要接入MMA SDK，SDK会在客户端采集设备相关的信息，并且随监测地址上报给第三方监测，支持的第三方监测主要以秒针为主。此类方式在技术上最难作假，也是很多广告主喜欢的方式。不过他也有自己的缺点
    - 视频媒体外，大部分中小媒体不愿意对接。
    - 此上报方式容易丢设备信息，从而造成较大的数据GAP（TA，频次等等）
    - 当期此方式获取设备信息过度频繁，会让接入媒体被工信部点名批评，一些较大的视频媒体也开始放弃此种方式
- API：媒体方无需接入MMA SDK，只需根据MMA的宏定义约定，通过替换监测链接代码中的宏参数，讲相关的设备信息随监测地址上报给第三方监测。

API代码均含有MMA标准参数，即`
__OS__, __IMEI__, __MAC__, __MAC1__, __IDFA__, __CAID__,__AAID__, __OAID__,__DUID__, __IP__, __UA__, __TS__ , __LBS , __GEO__
`

#### 参数优先级
- 对于 Android 系统，用户唯一标识的优先级顺序从高到低依次为：IMEI > OAID > MAC > AAID。
- 对于 iOS 系统，用户唯一标识的优先级顺序从高到低依次为：IDFA > CAID > MAC。

## 监测上报流程
监测上报中最重要为曝光，点击，唤起 三种监测
- 曝光监测：所有曝光监测地址，都由App端从协议中的曝光监测列表中提取所有曝光监测地址，逐条请求地址一达到上报曝光监测目的。
    + 串行逐条上报方式：等上一条上报结束再报下一条，实现简单，大部分媒体会采用此方式。
        - 会出现上报周期长，后面的链接丢数等情况。
        - 每条链接的超时时长短，监测服务器网络波动时，更容易丢数。特别https方式会增加丢数概率。
    + 并行上报方式：一次性发送所有链接，然后等待所有链接成功返回。
        - 上报周期短，但实现复杂度增加。
        - 整体超时时长短，但单条链接的超时时长相对串行方式会长很多。
- 点击监测：由于点击属于用户行为，会产生后续跳转行为，另外webview跳转只能是一条链接，估引出两种点击监测上报方式。**见以下点击上报流程图**
    + 同步点击上报：点击监测地址包在目标地址前面，通过http的302协议实现监测服务器收数以及成功跳转目标地址
        - 层级太深，中间一旦有一条监测地址失效超时，则无法成功跳转目标地址，从而造成目标地址丢失，所以不适合把所有的监测地址给串起来。
        - 不是所有点击监测服务都支持302协议。
    + 异步点击上报：针对上述方式的缺陷，引入异步点击上报方案，类似曝光监测，由APP端从协议中的点击监测列表中提取所有点击监测地址，逐条请求地址一达到上报点击监测目的。
        - 部分点击监测失效，不影响后续监测上报。
        - 无需所有点击监测服务支持302协议。
    + 由于点击监测有两条上报路径，不同的上报路径之间会有较大的概率出现数据GAP。
- 唤起监测：在点击跳转唤起APP时，媒体会请求一次唤起协议地址，告知监测服务器成功唤起APP。次监测非MMA标准协议中的一条，估只有部分直媒支持。
    + 受限于系统技术限制，唤起监测一般都在即将唤起APP之前上报，也就是它只能代表App尝试唤起目标App了，而非真实的成功唤起数。

## 无效流量
### 秒针无效流量规则
规则中的概念梳理

- ID: 代指IDFA, IMEI, OAID, AndroidID,MAC的任意一个非空值
- IMEI：可以代指 IMEI，IDFA，OAID 这3个在移动端（OTT除外）的核心设备ID
- DID(Device ID)：按优先级取数： m5 > m2 > m11 > m1a > m1 > m6a > m6 即 IDFA > IMEI > OAID > AndroidID > AndroidID1 > MAC > MAC1 (媒介从秒针那边得到的信息）
    - 不过用测试账号得到DID逻辑貌似Android ID在IMEI之前，在严格修正的SIVT规则中无视这种优先级变动的场景。
