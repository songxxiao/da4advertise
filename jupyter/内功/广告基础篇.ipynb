## 程序化广告
### 程序化广告
程序化广告是指利用技术手段，对整个数字媒体广告投放过程中的各个环节进行信息化，并通过技术手段衔接为一体的一种工具。

广告主可以程序化采购媒体资源，并利用算法和技术自动实现精准的目标受众定向，只把广告投放给对的人。

媒体可以程序化售卖跨媒体、跨终端（电脑、手机、平板、互联网电视等）的媒体资源，并利用技术实现广告流量的分级，进行差异化定价（如一线城市的价格高于二三线城市、黄金时段的价格高于其它时间段）。

#### RTB（Real-Time Bidding，实时竞价）
广告流量卖方通过程序化的方式将广告流量接入广告交易平台ADX，并设定底价，当用户浏览媒体内容页，有一个广告位需要展示广告时，卖方会将该广告曝光机会通过ADX向各程序化买家DSP发起竞价请求。各DSP根据对该广告曝光机会的评估背对背实时出价，ADX收到各个DSP的出价后，进行比价，找出出价最高的DSP，将出价最高的DSP的广告素材给到媒体进行展示，同时将竞价成功的结果返回到胜出的DSP，整个过程都是通过程序化的方式在100毫秒内完成。

#### PDB（Programmatic Direct Buy，私有程序化购买）
高大上的品牌广告主十分希望能通过程序化的手段对自己包段的黄金资源的流量进行程序化广告投放从而提升广告投放效率。于是就催生了介于传统排期和剩余流量RTB之间的变通模式PDB。IAB（美国互动广告局，The Interactive Advertising Bureau）称之为程序化合约（Automated Guaranteed）。

很多业内人士俗称保价保量模式。

#### PD（Preferred Deal，优先交易）
PD模式：流量卖方会为买方定一个相对较高的固定价（相对RTB的底价而言），广告请求来的时候若DSP判定这个流量是自己想通过PD的优先权来获取的，那么其会返回PD索要响应。该广告曝光机会就会优先展示该广告主的广告。这种模式也是买方卖方事前一对一约定的，只是卖方的库存是不保证的，价格是事前约定好并固定的。

####  DSP（Demand-Side Platform，需求方平台）
网络广告的程序化买方操作平台，通过这个操作平台的技术手段，买方可以根据自己的需求对目标人群每一次广告机会进行精准实时竞价购买。

两个必备的核心：一是以RTB的方式购买流量，二是支持需求方定制化的用户划分（user segmentation）

总结：是之前的投放系统更开放化的产物：对接了更多开放的资源，可以按需购买。


#### SSP（Supply-Side Platform，供给方平台）
SSP主要是指给流量卖方使用（接入）的平台。供给方平台，动态分配（dynamic allocation）媒体的资源到不同的消耗渠道上（Ad-Network，ADX等），提高媒体资源的变现能力。

总结：就是站在媒体方，帮助媒体去变现，提高媒体的变现能力。


#### ADX（Ad Exchange，广告交易平台）
ADX是衔接流量卖方（SSP），买方（DSP）的重要交易场所
广告交易平台。负责将媒体的流量以拍卖的方式售卖给DSP，类比于证券市场中的交易所。ADX一般以CPM结算。

总结：撮合平台，促成流量的供给方（Supply）和需求方（Demand广告主）之间的交易。


#### TD（Trading Desk）
TD是执行程序化购买，下单的一站式交易操作台。

随着分工不断精细化及专业化（比如，出现SSP，ADX，DSP，DMP等），大大增加了程序化广告下单执行及监控数据的复杂度，因此广告主（或代理公司）希望能一站式操控程序化交易。TD就是基于广告主这一需求出现的，其主要功能包括排期、下单、投放数据回收、报表展示等。

TD依据使用方不同可分为代理商、广告主内部、独立三种
- 代理商（ATD，Agency Trading Desk）：广告代理商公司内部专门为多个广告主提供服务，负责程序化广告购买投放的部门即Trading Desk。
- 广告主内部使用的Trading Desk：目前国内一些大型广告主正在建立自己的DMP+TD来管理自己的广告投放。
- 独立Trading Desk平台：主要是提供TD技术服务给广告公司或广告主。


#### DMP（Data-Management Platform，数据管理平台）
数据的收集和加工的平台，早期一般是提供第一方数据的管理（跟现在私有DMP类似）。把加工出来的受众标签，供客户分析用户画像，提供给投放端进行受众定向，从而进行变现。

总结：通过提供打标签的功能，用于投放定向，从而收取费用。

#### Ad Serving
Ad serving是站在广告主方面去运营已经属于广告主的流量（这就代表大部分的流量成本的流水是不过ad serving方的），做最优的搭配。功能是dsp功能的子集，只能做简单的地域定向、频次控制、人群定向（采购外部DMP）的功能，没有复杂的出价逻辑。主要应用在视频流量的PDB投放，或则display中banner部分的创意轮播（不同频次看不同创意的广告）。

代表公司有：reachmax，admaster，秒针等都有相关的产品

总结：是阉割版本的dsp，基本不保kpi，收费方式为服务费，按照整体的budget来收固定比例的服务费（通常少得可怜）。利润薄非常薄！

#### 第三方监测
一般广告主的广告投放不太可能以广告投放方的数据作为依据进行结算，这样自然就需要一个相对独立的第三方的监测数据分析方来提供结算依据。

- 国内品牌使用的较大的典型的第三方监测公司有秒针，RTBAsia，国双等
- 国际品牌使用的较大的典型的第三方监测公司有尼尔森，Sizmek等
- 专注移动监测的公司有TalkingData，友盟等



##  广告必备指标
### 广告考核指标
#### ROI（Return On Investment，投资回报率）
计算公式：`ROI =(Income - Cost)/Cost * 100% = 利润/成本 * 100%`

#### CPM（Cost Per Mille）
展现成本，或者叫千人展现成本，是指某广告在某一媒体某种广告位上被展示1000次的价格。

计算公式：`CPM = Cost/PV * 1000`

#### CPUV（Cost per Unique Visitor，每个独立访客的成本）
每个UV成本，即广告曝光中获取一个UV的价格。

计算公式：`CPUV = Cost/UV`

#### CPC（Cost Per Click，Cost Per Click-Through）
每个点击成本，即广告被点击一次对应的价格。

计算公式：`CPC＝Cost/Click`

#### CPD（Cost Per Day）
一个广告位一天的包断的价格

#### CPA（Cost Per Action，每次转化的成本）
一次转化（转化的定义可以是下载，激活，注册等）的价格，例如，游戏需要用户通过广告页面注册成为玩家，其CPA用公式表示即CPA＝Cost/注册数。

因此CPA还可具体分为
- CPD（Cost Per Download，每下载成本）
- CPI（Cost Per Install，每安装成本）
- CPS（Cost Per Sale，每销售成本）
- CPE（Cost Per Engagement，每互动成本）

计算公式：`CPA = Cost/Action = Cost/Click * Click/Action = CPC/CVR`

#### CPL（Cost Per Lead，每个留资的获取成本，每个潜在客户的获取成本）
获取一个潜在客户信息对应的价格，如汽车行业需要潜在客户通过广告页面填写表单预约试驾。

计算公式：`CPL = Cost/Lead`

#### CPH（Cost Per Hour）
受众对广告的浏览/互动时长的成本。根据广告可视度的概念可知，可视广告对受众浏览广告的面积和时长都有要求，因此CPH是评估可视广告的进一步延伸，它可以考核受众对广告的关注度、兴趣度和活跃度。

CPH指标特别适合品牌广告主，它在一定程度上代表了广告对受众品牌意识的影响力。

#### PV（Page View，网页浏览量）
对于单个站点，PV是网站各个网页被浏览的总次数（通常是指客户的落地页的浏览量），一个访客有可能会创造几十甚至更多的page views。 通常在投放中，PV也被当着广告被曝光的次数，这个时候就等同于Impression。

#### Impression（曝光）
广告被曝光的次数，通常以CPM为单位，一个CPM 1000次曝光

#### UV（Unique Visitor，独立访客数 ）
对于单个站点，一天中网站的独立访客数（以cookie或则设备ID为依据），一天内同一个访客多次访问网站，只算一个访客。 通常在投放中，UV是指看到广告的独立用户数（以cookie或设备ID为依据）。

#### Click（点击，点击数）
用户通过点击广告儿访问广告主的网页，称点击一次，点击数是评估广告效果的指标之一。

#### CTR（Click Through Rate，点击率）
广告点击次数占广告展示次数的百分比。

计算公式：`CTR ＝（Click/Impression）* 100%`。

#### CVR(Conversion Rate)
指广告的转化率，即广告转化次数占广告点击次数的百分比。转化次数可以是注册数、Lead数量、订单数等，根据实际业务约定。

计算公式：`CVR =（Coversion/Click）* 100%`

#### LR（Landing Rate，到达率，落地页的到达率）
用户点击广告后成功到达客户网站的比例。衡量从点击到客户到站之间的损耗的指标。 通常第一方人群的LR在70%左右，其他方人群的LR的数据区间很大，从10%到50%，具体要看投放系统和优化师的能力。

计算公式：`LR = Langding/Click * 100%`

#### 2nd-Click Rate（二跳率）
二跳是指浏览入口页后并在页面上产生的首次有效点击。二跳量是二跳的次数，二跳率是二跳量占进入该页面的总访问量的比例，可用于衡量用户的访问质量或该页面对用户的吸引力。

#### Bounce Rate（跳出率）
跳出的概念和二跳相对，跳出指浏览入口页之后没有在页面上产生有效点击就退出页面，跳出率是跳出量占进入该页面总访问量的比例，一般来说跳出率和二跳率的加和为1。

二跳率等于二跳量除以到达量（到达页面访问量）
```
2nd-Click Rate = 2nd-Click/Landing
Bounce Rate = Bounce/landing
(2nd-Click Rate) + (Bounce Rate) = 1
```


### 广告监测指标

#### Stable%（稳定率）
在PC端的投放，通常是通过cookie来标识一个唯一用户的，但是一部分cookie的使用者的存活时间太短（浏览器插件或则操作系统自动清cookie，或则用户手动清理cookie），这部分人群不能太好的计算频次和判断其用户的属性（计算TA），所以第三方会将其标记称非稳定人群（unstable），与之相对应的就是stable人群，不同监测公司对稳定cookie定义不同。

在PC端 Stable %的平均值在70%，***在Mob端一个设备对应的用户是固定的，所以不存在stable的概念，也就是Mob的Stable%均为百分之百。***

计算公式：`Stable% = (Stable PV)/(Total PV) * 100%`

#### Universe（TA Universe，目标网民总数）
常简称为Universe，指某个地区属于客户目标人群的网民的总数。比如：假如A市网民有2000万，其中18-35岁男网民（Target Audience）人口占25%，A市TA Universe 则为500万。

目标网民数，不是目标人口数，有些目标人群不一定上网的。

#### TA（Target Audience，目标受众）
目标受众可以是某一个人口群体，如年龄组、性别、婚姻状况、等等。常见受众有青少年、女性、单身、等等。目标受众也可以包括几个不同的人口群体，比如所有20到30岁的男性。营销过程也可以计划如何对待其他非目标群体。


#### TA%（Target Audience %，目标受众浓度，TA浓度）
TA浓度一般以百分比形式出现，代表目标受众用户的曝光数占总稳定曝光数的比。

计算公式：`TA% = (TA PV)/(Stable PV) * 100%`

#### Reach（目标人群到达率）
某个地区看到广告的目标受众占比该地区所有目标受众的比例。

一般来说，Reach指的是(1+Reach%)，指看过广告1次及1次以上目标受众占网民基数的比例。部分客户为了追求曝光的效果（因为频次会影响到对目标受众的影响力，频次太低影响力不够），也会看N+reach的指标（N通常为3），这里的N就是指目标用户观看的次数。

计算公式：`Reach = (Ta UV)/(TA Universe) * 100%`

#### Frequency（频次，展现频次）
广告展现平均频次，指一段时间内，平均每个人看到的广告的次数。

计算公式：`Frequency = PV/UV`

客户通常关心的是满足TA条件下人群的频次，所以计算公式计算的是TA Frequency，即目标受众的广告展现频次

计算公式：`TA Frequency = (TA PV)/(TA UV)`

#### IGRP（Internet gross rating points，互联网总收视点/毛评点/总收视率/毛收视率）

指的是目标人群从电视媒体的GRP（gross rating point，总收视点，主要统计一定期间内所有投放电视投放中目标人群收看广告情况）演变过来的概念。受媒体环境的变化，互联网的传播力、影响力日新月异，广告主把原来投放电视广告费用的一部分转投到互联网，所以互联网广告需要有对应的广告效果考量指标，就引用了电视媒体的GRP考量办法，发展到IGRP。

从下列公式看，这个指标就是一个次营销活动所触达到的目标用户的观看广告的次数。客户当然希望这个值越大越好。这个指标的缺点为重复覆盖的现象并没有在指标上有体现，比如1个目标用户看1000次广告和1000个目标用户每个人看1次广告，得到的iGRP的值是一样的，但是实际传播的效果差异很大。

计算公式：`IGRP = Reach * Frequency * 100%`
