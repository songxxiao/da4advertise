# 移动广告
目前移动端已经超越PC端成为主流，占领了用户的大部分时间碎片。

移动APP媒体主要采用SDK、API的方式进行技术对接。
- SDK：广告的展示及用户手机的相关数据的获取均由SDK代码完成。SDK采用自己的设备ID规范，不需双方统一设备ID规范。采用这种方式由于媒体卖方丧失了对流量的控制权，估稍大一些的媒体卖方一般均不支持该模式。
- API：API主要采用的是服务端接口对接（S2S）的方式，大都会采用基于OpenRTB标准协议进行定制的方式。双方需要遵守统一的设备ID规范。这种方式由于媒体卖方可对流量进行控制，估是常见的技术对接方式。

## 移动端特有的一些问题
- 移动端APP上主要以设备ID来标识一个人，若获取不到设备ID，则无法监测频次等数据。目前iOS14无法获取IDFA就是广告行业的一大挑战
- APP中用户的行为内容较PC更难被抓取，若不使用SDK，或无媒体的支持，则很难获取上下文数据。
- 移动端地域GAP较PC端更严重。
- 移动端屏幕小，易误点击及户外4G/5G上网网络不稳定等因素。


## APP广告的常见形式
### 开屏
- 开屏广告又称全屏广告（Full Screen Ads），几乎在常用的app上都能看到，可以以图片的形式、视频、Flash等形式加载。用户首次进入app时，将会出现启动页，当app后台运行再次进入时，启动页将不会出现。
- 常出现位置：首页启动页
- 优点：
    1. 合理利用资源，等待app加载时呈现内容
    2. 当用户刚打开启动页时，直接呈现广告内容，能够更好的刺激用户记忆。
- 缺点：部分Flash安装包加在缓慢，影响用户体验。

![开屏](../imgs/开屏.png)

### 信息流
- 信息流广告（Feeds Ads），这种广告伴随着信息而出现，用户在浏览信息时会不经意间浏览到广告。
- 常出现位置：以TimeLine信息为主的内容列表里。
- 优点：不影响用户操作行为。
- 缺点：内容定位不精准会让用户产生厌恶情绪。
![信息流](../imgs/信息流.png)

### 视频贴片
- 视频广告针对的用户群体分为VIP用户和普通用户，VIP用户在购买VIP业务后能够直接跳过广告，普通用户则需要先把广告看完才能看后面的内容。所以，这种广告方式收入渠道分为两种：一种是会员业务，另一种是视频广告。
- 常出现位置：播放类app内容开头（前贴片），中间（中贴片），结尾（后贴片）
- 优点：以内嵌的形式植入广告，不增加额外的内容板块。
- 缺点：普通用户（不愿意付费购买VIP业务）长时间观看会影响用户体验。
![贴片](../imgs/贴片.png)

### Banner
- Banner广告又叫横幅广告、通栏广告、广告条。
- 常出现位置：顶部、底部。
- 优点：展示更直观，能快速吸引用户注意。
- 缺点：影响用户体验，对内容观看造成一定的遮挡，易造成用户反感。
![Banner](../imgs/Banner.png)

### 插屏
- 一般情况下插屏广告会出现在用户第一次点击某个功能页时弹出，显示需要提示的具体内容。
- 常出现位置：首页、未点击的功能页。
- 优点：视觉冲击力强、定位更精准、效果显著
- 缺点：会暂时打断用户操作行为，影响用户体验。
![插屏](../imgs/插屏.png)

### APP没安装怎么办？
一般会采取以下两种方式
- 加一个Moible Web的中间页，在该中间页上安置JS代码来判断，通过Deep Link尝试调用手机APP，若手机上APP没安装，则跳转到APP下载页面引导用户安装。
- 广告主会提供Deep Link，H5内容页一组连接，媒体端判断，若手机上APP没安装，则打开H5页面，否则调用Deep Link尝试调用手机APP
    - 这种方式会存在APP安装，但Deep Link吊起失败的情况。
    
### Universal Link
很多APP服务提供方已经将提供普通网站URL的方式作为中间页载体，从而便于传播和使用。于是在iOS上引入了一个新的规范Universal Link。相当于系统级别实现了一个中间也载体功能

- 当APP未安装时则停留在当前H5页面
- 当APP安装时，则会自动唤起APP
- 唤起APP时不会弹跳转提示框

## 设备ID
当广告在媒体上曝光后，如何知道曝光给了多少人，每个用户的平均曝光次数是多少，最大的用户曝光了几次？这一系列问题实际上可以抽象成一个问题，那就是如何识别每次曝光是给哪个真人的？
    
在日常工作生活中常用身份证件号（身份证，护照，驾驶证等各种证件），手机号来识别一个人。在医学上常用DNA鉴定的方式来识别一个人。但在媒体流量中，不光获取这些信息的难度很大，单从大家都关注的用户隐私角度来看，就不能用这些信息作为识别真人的标准。

在从互联网角度来看，每家APP都会有自己的账号体系（比如QQ号，微信号，支付宝账号等），那么这些信息是否可以用来识别真人呢？理论上是可以的，不过依旧存在以下问题1. 用户隐私 2. 各家APP自成体系，无法跨平台识别同一个人。于是广告体系中也不会用这些信息来识别真人。

可用在广告体系中标记真人的信息需要尽量满足以下几点1. 尽量脱敏，标识是一个独立人，但不能有更多的用户信息 2. 可跨平台使用，也就是尽量平台无关 3. 尽可能的稳定，不能在极短时间内发生变化。手机的设备ID恰好就能满足以上几点要求，实际上设备ID不只是一种单一标记，它是一系列编码的统称，包含部分硬编码（比如IMEI，MAC），部分手机系统生成的软编码（IDFA，OAID，CAID）。

以下对常用的一些设备ID做个介绍

| 英文缩写   | 中文名称                              | 长度     |
| ---------- | ------------------------------------- | -------- |
| IMEI       | 国际移动设备识别码（安卓10以下专供）  | 15~17位  |
| IDFA       | iOS广告ID                             | 最长64位 |
| OAID       | 匿名设备标识符(安卓10以上专供)        | 最长64位 |
| CAID       | 广协针对iOS14无法获取IDFA提供的标识符 |          |
| AAID       | 阿里针对iOS14无法获取IDFA提供的标识符 |          |
| MAC        | 网卡设备硬件识别码                    | 48位     |
| Android ID | Android设备的唯一识别码               | 最长16位 |

### IMEI
当前一般只有安卓10以下（不包含安卓10）媒体可以获取，并且作为安卓10以下设备唯一性标识，使用中一般会做MD5脱敏加密。

目前厂商在媒体无法获取IMEI号是，会返回一个默认值，这点需要注意

```
'DD4B21E9EF71E1291183A46B913AE6F2' -- md5('00000000')
'80F54F6966AB4A4CD3C913E653CC4180' -- md5('812345678912343') OPPO默认填充值
'6F86A101B1DDA5F50BCA7E110E0C0452' -- md5('865407010000009') VIVO默认填充值
'CFCD208495D565EF66E7DFF9F98764DA' -- md5('0')
'4B997118E2DB4480B95D28CC741B3917' -- md5('812345678912345') OPPO默认填充值
'5284047F4FFB4E04824A2FD1D1F0CD62' -- md5('000000000000000')
'AD921D60486366258809553A3DB49A4A' -- md5('unknown')
```

安卓10（包含安卓10）媒体几乎无法获取IMEI，于是才有了中国移动安全联盟联合各大手机设备商提供了系统级别的匿名设备标识符（OAID）。


### IDFA
iOS一般是限制MAC，IMEI的获取权限，并且系统级别提供了一个IDFA，作为设备唯一性标识使用，使用中使用明文。
从iOS 14（2021-02-01）开始，iOS开始加大IDFA获取权限的控制，每个APP都需要单独得到用户的授权才能获取IDFA，也就是会造成大部分媒体中无法获取IDFA


### OAID
由于Android Q(10)版本功能限制应用获取IMEI， DEVICE ID等标识，中国移动安全联盟MSA联合各大手机设备商场，共同制定统一的匿名设备标识符（OAID）来替换IMEI，MAC地址。

抛开OAID的商务场景不说，光从技术层面来说，它实际上就是一个UUID，一个通过当前时间，随机数，名称，硬件信息等信息通过一组算法生成一个64位长度的随机值，按照定义，它应该是在新手机第一次开机时，由系统层面生成，用户在后期可以随时重置这个值。

### CAID，AAID
从iOS 14（2021-02-01）开始，iOS开始加大IDFA获取权限的控制，每个APP都需要单独得到用户的授权才能获取IDFA，也就是会造成大部分媒体中无法获取IDFA，针对此种情况，目前的现状是中国广协提供CAID（收费），阿里提供了自己的AAID，但这些新的设备ID推进工作还在进行中，由于非系统界别，需要每家APP单独接入，推进工作目前较慢

### MAC
在移动端下一般单指无线网卡的硬件识别码，使用中一般会做MD5脱敏加密。

MAC值的格式为"xx:xx:xx:xx:xx:xx"，计算MD5时会有两种方式，一种是保留':'，一种是去掉':'，于是相同的MAC值会出现两种加密值

无论是安卓还是iOS，大部分情况下都是无法获取MAC地址了，但系统一般会返回一个默认值，目前已知的有以下几个值

```
'E3F5536A141811DB40EFD6400F1D0A4E'  -- md5('020000000000')
'35B9AB5A36F3234DD26DB357FD4A0DC1'  -- md5('000000000000')
'528C8E6CD4A3C6598999A0E9DF15AD32'  -- md5('00:00:00:00:00:00')
'0F607264FC6318A92B9E13C65DB7CD3C'  -- md5('02:00:00:00:00:00')
'D41D8CD98F00B204E9800998ECF8427E'  -- md5('')
```

### Android ID
这是安卓系统下的一个系统级别的设备唯一标识码，用户可以重置它，目前真实情况下并不怎么使用这个值，使用中一般会做MD5脱敏加密。

### 设备ID
在真实的使用中，为了简化问题，需要用以上各种设备ID组装成一个唯一的设备ID。虽说这些设备ID各自都是针对不同场景出现的，但还是会有重合的场景出现，那么针对这些重合的场景，如何使用它们呢。一般是按以下的优先级来取其中一个值放到唯一的设备ID中（IDFA > IMEI > OAID > CAID > AndroidID > MAC）


## 设备信息
移动端设备中，除了标识唯一性的设备ID外，实际上还有一些非敏感信息可以使用，比如品牌，机型等，这些信息用来做定向投放。

- APP名称：媒体流量来源APP的名称
- APP版本：一般查问题时需要看下
- 系统：os还是android
- 系统版本
- 屏幕分辨率
- 品牌：华为，vivo, oppo, 小米，荣耀，红米等
- 机型：具体的机器型号
- UA(User Agent):中文名为用户代理，简称 UA，它是一个特殊字符串头，使得服务器能够识别客户使用的操作系统及版本、浏览器及版本、浏览器语言、浏览器插件等。
- 经纬度：
    - 维度Latitude的范围是：-90 到 +90
    - 经度Longitude的范围：-180 到 +180
- IP：IP地址是用来识别网络上的设备，由网络运营商（电信，移动等）分配，它是一个比较底层的协议，无法伪造改变
    - 目前IP可以使用IP广协库来定位到城市信息，一般用于地域定向投放
    - IP目前分IPv4跟IPv6两个版本，当前还是以IPv4为主，IPv6还为流行
    - IPv4格式：225.225.225.225
    - IPv6格式：CDCD:910A:2222:5498:8475:1111:3900:2020
- 网络类型：WIFI，3G，4G，5G

### User Agent(UA)

UA在客户端上报监测的时候带在http header，在移动端UA一般会有三大类其中第一，二类为系统的标准格式，第三类则是一些APP自行组装

- Dalvik虚拟机，是Google等厂商合作开发的Android移动设备平台的核心组成部分之一。此类格式一般出现在三方异步的曝光，点击监测中，不应该出现在同步点击监测中

```
Dalvik/2.1.0 (Linux; U; Android 9; BKL-AL20 Build/HUAWEIBKL-AL20)
Dalvik/2.1.0 (Linux; U; Android 9; HMA-AL00 Build/HUAWEIHMA-AL00)
```
- Mozilla，中文名称摩斯拉。现在市面上大部分浏览器的ua都是以Mozilla开头，由于我们在APP点击了广告，就会去打开相关页面，这时候需要浏览器来打开页面。此类格式可以出现在任意监测模式中

```
Mozilla/5.0 (Linux; Android 8.0.0; HUAWEI NXT-TL00 Build/HUAWEINXT-TL00; wv) AppleWebKit/537.36 (KHTML, like Gecko) Version/4.0 Chrome/66.0.3359.126 MQQBrowser/6.2

Mozilla/5.0 (Linux; Android 6.0.1; Coolpad 5370 Build/MMB29M; wv) AppleWebKit/537.36 (KHTML, like Gecko) Version/4.0 Chrome/55.0.2883.91 Mobile Safari/537.36
```

- APP自己根据设备信息拼接出来，比如iQiYi, 优酷等都是自己独有的格式。此类格式一般出现在三方异步的曝光，点击监测中，不应该出现在同步点击监测中

```
QIYIVideo/11.12.0 (iOS;com.qiyi.iphone;iOS13.6;iPhone8,1) Corejar
OTTSDK;2.2.0.29;Android;4.4.2;LCD-xxCAE5A-C
```

### 经纬度
在做LBS的应用的时候考虑到了这样一个问题，经纬度后面的小数的位数对于精度影响有多少？

- 维度Latitude的范围是：-90 到 +90。在纬度相等的情况下：
    - 经度每隔0.00001度，距离相差约1米；
    - 经度每隔0.0001度，距离相差约10米；
    - 经度每隔0.001度，距离相差约100米；
    - 经度每隔0.01度，距离相差约1000米；
    - 经度每隔0.1度，距离相差约10000米。
- 经度Longitude的范围：-180 到 +180。在经度相等的情况下：
    - 纬度每隔0.00001度，距离相差约1.1米；
    - 纬度每隔0.0001度，距离相差约11米；
    - 纬度每隔0.001度，距离相差约111米；
    - 纬度每隔0.01度，距离相差约1113米；
    - 纬度每隔0.1度，距离相差约11132米。
