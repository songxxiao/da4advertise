## Easyreport
### fancy项目（临时表）
1. 仪表盘id：268 秒针SIVT监控报表  
用于查看挂有秒针监测的订单投放中sivt的情况

2. 仪表盘id：82  独代媒体估量仪表盘  
趣头条、韩剧TV、人人视频独代媒体各维度数据，更新周期较长，一般由数据分析根据媒体对外汇报数据进行修改
### fancyreport项目（长期维护）
#### 常用表  
1. 仪表盘id：265 DSP-5分钟多维度报表  
DSP实时报表，维度包含：  
天、月、小时、5分钟、广告主、AMO订单、投放、投放计划、投放组、点位、AMO排期、ADX、广告位、操作系统、订单标签、省份、城市、ABtset、自定义广告主维度  
可根据投放相关数据按上述维度将数据呈现。  
自定义维度使用：维度勾选自定义维度，并在自定义广告组维度填入json  
{"字段名称1"：{"广告组名称1":"分类1","广告组名称2":"分类2","广告组名称3":"分类3"},"字段名称2"：{"广告组名称4":"分类4","广告组名称5":"分类5","广告组名称6":"分类6"}}  
eg：{"广告位":{"开屏":"开屏1","信息流":"信息流2"}, "城市":{"杭州":"杭州","深圳":"深圳","上海":"上海"}}  
注意事项：该表为投放维度表，对与指标请求数和参竞数在汇总情况下会有放大。  


2. 仪表盘id：266 FTX实时消耗报表  
FTX实时报表，维度包含：  
天、月、周、小时、5分钟、FTX子订单、FTX父订单、AMO订单、ADX名称、ADX广告位、FTX广告位、DSP名称、订单标签、城市、省份、操作系统、终端、USOP媒体、USOP点位、ABtset
可根据投放相关数据按上述维度将数据呈现。  
注意事项：该表为订单维度表，对与指标请求数和参竞数在汇总情况下会有放大。  


3. 仪表盘id：272 FTX可见性报表  
针对视频素材播放情况报表，在挂有完整监测的订单条件下，可查看一帧播放、5s播放、15播放等数据。

4. 仪表盘id：274 媒体广播量报表
媒体维度报表，请求、响应数据不会放大。
维度包含：ADX名称、FTX广告位、ADX广告位、广告位类型、广告位尺寸、城市、省份、操作系统、终端、包名  


5. 仪表盘id：284 媒体快速估量  
主要用于查看媒体用户数信息


6. 仪表盘id：311 FTX频次报表  
针对挂有频控id的子订单、父订单、频控id查看其频次信息  

7. 仪表盘id：378 FTX-订单流量抓包  
千分之一抽样，可根据任意关键词查找流量中带有关键词的json，流量信息包括媒体->FTX，FTX->DSP,DSP->FTX,FTX->媒体

8. 仪表盘id：278 DSP-旺财系统报表  
根据旺财离线表所出数据  

9. 仪表盘id：405 新旺财报表  
将旺财数据直接用adid关联的新旺财数据，与前表不兼容  
  
10. 仪表盘id：357 DSP-Render系统报表
