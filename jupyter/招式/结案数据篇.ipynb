# 结案数据篇
## 背景
一个订单执行结束后，需要给客户反馈一份结案数据，结案数据基本上就是把曝光点击按单维度点位，年龄，性别，兴趣等以一定的比例分摊下去。

注意事项
- 以下几个维度必须按实际真实数据来出，不可美化。
    + 小时，地域，品牌，系统
- 其它维度可以依据AM，运营的需求使用工具做一定的比例美化。

## 执行要求
执行过程中注意三点1. 充分沟通，胆大心细。2. 充分沟通，胆大心细。3. 充分沟通，胆大心细。

- 动手前一定要先做好AM，运营，数分三方沟通，确认清楚需求明细，避免沟通GAP造成反复返工。
- 执行中一定要做到胆大心细，并及时反馈疑问点，沟通清楚。
- 执行后一定要对数据负责。
    + 曝光、点击要跟总数一致。
    + 维度比例满足沟通定下的要求。

## 执行工具
执行工具主要有三宝：EasyReport, Excel, 蛋饼，以下核心讲解下蛋饼工具的使用

### 蛋饼
[蛋饼工具](http://amo.fancydsp.com/tools/dataProcessing)是研发这边帮忙开发一个专用用于拆分数据到单维度上的一个工具，可以在界面中分别调整每个维度中值项的比例，以最终得到想要的美化结果数据，同时又能确保求和数据与原始的曝光数，点击数相等。

蛋饼工具以一个excel模板作为输入，模板核心包含两部分数据1. 需要拆分的原始曝光，点击数 2. 需要拆分到的维度，以及维度值。最后可以下载一份拆分好的excel格式数据。

> 蛋饼的命名来源于摊蛋饼，拆分数据就如果摊蛋饼一样，把一个圆圆的元数据摊到各个维度上。

#### 蛋饼模板
> 点击下载[蛋饼模板](../文档/蛋饼模板.xlsx)

##### 原始数据(rawdata)
![蛋饼模板rawdata](../imgs/结案/蛋饼模板rawdata.png)

- 原始数据必须放在excel的**第一个工作表(sheet)**中，命名不重要，表格格式不重要，建议按照模板命名来
- 严格按模板来，表格必须以第一行第一列(A1)开始。
- 第一行为目标标题，命名看着很学生，但是不能修改，必须保留p, l, d, dc等命名
- 第二行为输出数据标题，除了第二列根据需求修改，其它部分都保留模板定义
- 第二行第二列为需要摊分的维度`sex/age/category`
    - 标记需要摊分的维度，以'/'切分，每一个值代表一个维度，需要有一个对应的工作表(sheet)来输入对应维度的值项。
    - 可根据需要增减维度，并不是只能分三类。
    - 命名无强制约定，只需要与对应的维度工作表(sheet)的命名保持一致即可，建议命名有意义的名称。
- 第三行开始为需要拆分的原始数据，输入部分只需要填人群，展现，点击即可
    - 技术上支持输入多个人群信息，也就是可以有第四行，第五行数据，每一行代表一个人群数据，但实际中不太建议这么使用，原因为这样子每个人群最后在维度上摊出来的比例都比较接近，不一定满足需求。

##### 维度值
![蛋饼模板dim](../imgs/结案/蛋饼模板dim.png)

- 每一个维度需要单独建一个工作表(sheet)，**注意不要占用第一个工作表**，**命名很重要**，表格格式不重要
- 严格按模板来，表格必须以第一行第一列(A1)开始。
- 第一行为类名，并且需要做到1. 原始数据中的维度名 2. 工作表名 3. 首行类名 三个值必须保持一致相同。
- 从第二行开始为维度的选项值，中间不可有空行。

#### 蛋饼系统
操作步骤

1. 按照需求修改蛋饼模板。
2. 登录[蛋饼系统](http://amo.fancydsp.com/tools/dataProcessing)并点击`上传`把数据模板上传到蛋饼系统中
2. 逐个维度调整比例，常规做法为
    + 分布类型选择"固定分布"，波动类型选择"随机联动"，级联类型选择"波动随机"。其它类型可自己研究并补充到这里
    + 在分布字段处点击配置，然后根据业务需求拖动维度值的比例
3. 预览数据确认结果，并且点击`下载`来下载结果数据
